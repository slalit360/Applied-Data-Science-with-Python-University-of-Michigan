```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
#  4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 100)
x2 = np.random.gamma(2, 1.5, 100)
x3 = np.random.exponential(2, 100)+7
x4 = np.random.uniform(14,20, 100)

# plot the histograms
plt.figure(figsize=(9,3))
plt.hist(x1, normed=True, bins=80, alpha=0.5)
plt.hist(x2, normed=True, bins=80, alpha=0.5)
plt.hist(x3, normed=True, bins=80, alpha=0.5)
plt.hist(x4, normed=True, bins=80, alpha=0.5)

plt.axis([-7,21,0,1])

plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
plt.text(x3.mean()-1.5, 0.5, 'x3\nExponential')
plt.text(x4.mean()-1.5, 0.5, 'x4\nUniform')
```

```python
import matplotlib.animation as ani
%matplotlib notebook
n = 100
title = ['Normal', 'Gamma', 'Exponential', 'Uniform']
col = ['blue', 'orange', 'green', 'red']



def update (curr):
    if curr == n:
        a.event_source.stop()
        
    plt.cla()
    bins = np.arange(0, 8, 0.5)
    
    for i in range(0, len(axs)):
        axs[i].hist(plts[i][:curr], bins=bins, color=col[i])
        axs[i].axis([-1, 8, -0.05, 50])
        axs[i].set_title(title[i], fontsize=11)
        axs[i].set_xticks(np.arange(0, 10, 2))
        axs[i].set_yticks(np.arange(0, 60, 10))

    ax1.tick_params(labelleft='on',labelbottom='on')
    ax2.tick_params(labelleft='on', labelbottom='on')
    ax3.tick_params(labelleft='on', labelbottom='on')
    ax4.tick_params(labelleft='on', labelbottom='on')

#redefine histogramns \n",
x1 = np.random.normal(loc=3.0, size=100)
x2 = np.random.gamma(shape=1.5, size=100)
x3 = np.random.exponential(size=100)
x4 = np.random.uniform(0, 6, 100)

plts = [x1, x2, x3, x4]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=True)

axs = [ax1, ax2, ax3, ax4]

a = ani.FuncAnimation(fig, update, interval=10)
```
